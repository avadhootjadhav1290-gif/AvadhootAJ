```python
# Q5 Which district has the highest Literacy Rate (%) 
# Literacy Rate (%) means the percentage of people who can read and write in a population.

c.loc[c['Literacy Rate (%)'].idxmax(),['District Name','Literacy Rate (%)']]
```

## Questions for Grpha Plotting

```python
# How are the 36 districts of Maharashtra distributed across its Administrative Divisions?
sns.countplot(data=c, y='Administrative Division')
```

```python
# Q2 What is the relationship between district population and area, and how do Sex Ratio and Population Density influence this relationship?
sns.relplot(data=c, x='Area (in sq. km)', y='Population (Census 2011)',kind='scatter', hue='Sex Ratio', size='Population Density (per sq. km)')
```

```python
# Q3 What is the probability density distribution of the Sex Ratio (females per 1000 males) across the districts?
sns.kdeplot(data=c, x='Sex Ratio', fill=True, color='red')
```

```python
sns.pairplot(c,hue='District Name')
```

```python
# Q5 What is the distribution and spread of Urban Population Percentage across different Administrative Divisions?
sns.boxplot(data=c, x='Urban Population (%)', y='Administrative Division',palette=['red','blue','green','yellow','pink','gray'])
```
